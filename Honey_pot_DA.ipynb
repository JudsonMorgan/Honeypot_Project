```python
path_list = []
data_list = []
name = []
def load_data(path):
    for i in range(23):
        path_list.append(path.replace(path[29], str(i)))
    for i in range(1,23):
        if i == 5 or i == 6:
            continue
        honeytrap = pd.read_json(path_list[i],lines=True)
        data_list.append(honeytrap)
load_data(c)
```

```python
list(list(honeytrap_data.attack_connection.values)[6].values())
```

```python
def connection(honeytrap):
    dst_ip = []
    dst_port = []
    src_ip = []
    src_port = []
    protocol = []
#     event_properties = list(col)
    for i in range(honeytrap.shape[0]):
        col = list(list(honeytrap.attack_connection.values)[i].values())
        protocol.append(col[0])
        dst_ip.append(col[1])
        dst_port.append(col[2])
        src_ip.append(col[3])
        src_port.append(col[4])
    honeytrap_data = pd.DataFrame(list(zip(dst_ip,dst_port, src_ip,src_port,protocol)), columns = ['dst_ip','dst_port', 'src_ip','src_port','protocol'])
    return honeytrap_data
honeytrap_data = connection(honeytrap_data)
```

```python
import requests


def get_ip():
    response = requests.get('https://api64.ipify.org?format=json').json()
    return response["ip"]


def get_location():
    ip_address = '181.219.150.147'
    response = requests.get(f'https://ipapi.co/{ip_address}/json/').json()
    location_data = {
        "ip": ip_address,
        "city": response.get("city"),
        "region": response.get("region"),
        "country": response.get("country_name")
    }
    return location_data


print(get_location())
```
